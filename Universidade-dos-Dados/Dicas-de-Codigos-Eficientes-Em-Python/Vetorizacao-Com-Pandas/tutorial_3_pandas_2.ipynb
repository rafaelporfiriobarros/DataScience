```python
def low_values_red(value):
    color = 'red' if value < 92 else 'black'
    return 'color: %s' %color

fifa19[['Overall', 'Potential']].head(10).style.applymap(low_values_red)
```

```python
def highlight_max(s):
    is_max = s == s.max() # checa esse booleano s == s.max(), e depois guarda em is_max
    return ['background-color: yellow' if v else '' for v in is_max]

fifa19.head(10).style.apply(highlight_max, subset=['Age', 'Overall', 'Potential'])
```

```python
def substitui_lbs(value):
    """ funcao para tratar campo peso """
    x = value.replace('lbs', '')
    return int(x)

df['Weight_New'] = df['Weight'].apply(substitui_lbs)
df.head()
```

```python
df[['Weight_New']].info()
```

```python
# Convertendo campo salário

def converte_wage(value):
    value_2 = value.replace('€', '')
    if 'M' in value_2:
        value_2 = value_2.replace('M', '')
        return float(value_2)*1000000
    elif 'K' in value_2:
        value_2 = value_2.replace('K', '')
        return float(value_2) * 1000
    
df['Wage_Numeric'] = df['Wage'].apply(converte_wage)
df.head()
```

```python
df['avg_GK'] = (df['GKDiving'] + df['GKHandling'] + df['GKKicking'] + df['GKPositioning'] + df['GKReflexes']) / 5
df.head()
```

```python
df['Age_Class'] = np.where(df['Age'] > 30, 'Old', 'Young')
df.head()
```

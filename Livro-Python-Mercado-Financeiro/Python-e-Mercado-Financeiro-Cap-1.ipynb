## 1.4 OPERAÇÕES COM ELEMENTOS DAS LISTAS NO CONSOLE

```python
# ibov[2:4]: do elemento de índice 2 ao índice 3

ibov[2:4]
```

```python
# ibov[1:]: do elemento de índice 1 até o último

ibov[1:]
```

```python
# ibov[:3]: do elemento inicial (índice zero) ao elemento de índice 2 (pois é 3 − 1)

ibov[:3]
```

```python
# ibov[0:5:2]: do elemento inicial ao último saltando de 2 em 2

ibov[0:5:2]
```

```python
# ibov[−3:]: seleciona os 3 últimos elementos da lista

ibov[-3:]
```

```python
# ibov[:−3]: seleciona os 2 primeiros elementos da lista (pois é −(3 − 1))

ibov[:-3]
```

```python
print(retorno)
```

```python
t = ny.arange(0,19)

fig.plot(t, retorno)

fig.title('retorno')

fig.xlabel('tempo')
fig.ylabel('retorno')
```

## EXEMPLO 1.2

```python
import math
import statistics as st
```

```python
media = st.mean(retorno)
```

```python
desvio = st.pstdev(retorno)
```

```python
otimista = media + 2 * desvio / math.sqrt(19)
```

```python
pessimista = media - 2 * desvio / math.sqrt(19)
```

```python
print(otimista, pessimista)
```
